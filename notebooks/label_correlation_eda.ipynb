# Label Correlation EDA
Compute pairwise correlations between the 6 emotion labels using train and validation splits.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train_path = '/home/dinithi/Documents/EMI_Challenge/data/splits/train_split.csv'
valid_path = '/home/dinithi/Documents/EMI_Challenge/data/splits/valid_split.csv'

train_df = pd.read_csv(train_path)
valid_df = pd.read_csv(valid_path)
df = pd.concat([train_df, valid_df], ignore_index=True)

label_cols = ['Admiration', 'Amusement', 'Determination', 'Empathic Pain', 'Excitement', 'Joy']
missing = [col for col in label_cols if col not in df.columns]
if missing:
    raise ValueError(f'Missing label columns: {missing}')

corr = df[label_cols].corr(method='pearson')
print('Correlation matrix:')
display(corr.round(3))
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', square=True)
plt.title('Emotion Label Correlation (Train + Valid)')
plt.tight_layout()
plt.show()

# strongest off-diagonal pairs
pairs = corr.where(~corr.index.to_series().eq(corr.columns, axis=0)).stack().reset_index()
pairs.columns = ['label_1', 'label_2', 'corr']
pairs['abs_corr'] = pairs['corr'].abs()
pairs = pairs[pairs['label_1'] < pairs['label_2']].sort_values('abs_corr', ascending=False)
print('Top correlated label pairs:')
display(pairs.head(10)[['label_1', 'label_2', 'corr']])
```
